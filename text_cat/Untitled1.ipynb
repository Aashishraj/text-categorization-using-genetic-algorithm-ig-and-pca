```python
import pandas as pd
```

```python
data = pd.read_csv('dataset_6cat1.csv',encoding='latin_1')
```

```python
from nltk.corpus import stopwords
```

```python
from nltk.tokenize import word_tokenize
```

```python
import nltk
```

```python
stop_words=set(stopwords.words("english"))        
```

```python
for i in range(0,len(data.doc_text)):
            sen=data.doc_text[i]
            words=word_tokenize(sen)
            f=[w for w in words if not w in stop_words]
            s=" "
            o=s.join(f)
            data.doc_text.replace(data.doc_text[i],o,inplace=True)
data.doc_text
        
```
